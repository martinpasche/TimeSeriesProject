| Metric                              | Formula                                                                                                       | Meaning                                                                                                          | Utilization                                                                                                     |
|-------------------------------------|---------------------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------------|----------------------------------------------------------------------------------------------------------------|
| **Pearson Correlation (`pearsonr`)** | $ r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}} $     | Measures the **linear relationship** between two variables. A value of 1 indicates perfect positive correlation, -1 perfect negative, and 0 no correlation. | Use when you suspect **linear relationships** between variables. Requires continuous, normally distributed data.|
| **Spearman Correlation (`spearmanr`)** | $ \rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} $                                                              | Measures **monotonic relationships** by comparing the ranks of the data instead of raw values. | Useful when variables have **non-linear** but monotonic relationships. Handles outliers better than Pearson. |
| **Kendall's Tau (`kendalltau`)** | $ \tau = \frac{C - D}{\binom{n}{2}} $, where $ C $ and $ D $ are the counts of concordant and discordant pairs, respectively. | Measures the ordinal association between two variables based on concordant (C) and discordant (D) pairs. | Works well for **ranked data** or small datasets. Sensitive to ties and outliers.                            |
| **Chi-Square Test (`chi2_contingency`)** | $ \chi^2 = \sum \frac{(O - E)^2}{E} $, where $ O $ are observed frequencies and $ E $ are expected frequencies. | Tests the independence of two categorical variables by comparing observed (O) and expected (E) frequencies. | Use for **categorical variables** in contingency tables. Helps determine if variables are dependent.         |
| **Mutual Information (`mutual_info_regression`/`mutual_info_classif`)** | $ I(X; Y) = \sum_{x, y} P(x, y) \log \frac{P(x, y)}{P(x) P(y)} $                                             | Measures the general **dependency** (linear or non-linear) between two variables. Returns the amount of information one variable provides about another. | Use for **non-linear relationships**. Particularly useful in **feature selection** for regression/classification tasks. |
| **Normalized Mutual Information (`normalized_mutual_info_score`)** | $ NMI = \frac{2 I(X; Y)}{H(X) + H(Y)} $, where $ H(X) $ and $ H(Y) $ are the entropies of $ X $ and $ Y $. | Scales mutual information to range [0, 1], where 0 indicates independence and 1 indicates perfect dependence. | Useful for comparing clusterings or evaluating dependency between variables in a **bounded scale**.          |

---

### Explanation of Symbols:
- $ x_i, y_i $: Individual observations for variables $ X $ and $ Y $.
- $ \bar{x}, \bar{y} $: Mean of $ X $ and $ Y $.
- $ d_i $: Difference in ranks for observation $ i $ in Spearman correlation.
- $ C, D $: Concordant and discordant pairs in Kendall's Tau.
- $ O, E $: Observed and expected frequencies in Chi-Square test.
- $ P(x, y) $: Joint probability of $ x $ and $ y $.
- $ P(x), P(y) $: Marginal probabilities of $ x $ and $ y $.
- $ H(X), H(Y) $: Entropies of $ X $ and $ Y $.

### Usage Guide
1. **Linear relationships**: Use **Pearson**.
2. **Monotonic relationships**: Use **Spearman** or **Kendall**.
3. **Categorical independence**: Use **Chi-Square Test**.
4. **Non-linear dependencies**: Use **Mutual Information**.
5. **Bounded dependency comparison**: Use **Normalized Mutual Information**.


# Useful Feature Correlation Statistics for Time Series

| **Statistic/Method**           | **Purpose**                                                                                       | **Strengths**                                                                                      | **When to Use**                                                                                             |
|--------------------------------|---------------------------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------------------|
| **Pearson Correlation**         | Measures **linear correlation** between two variables.                                           | Simple to compute; good for stationary data with linear relationships.                           | Use if both series are stationary and have a clear linear relationship.                                   |
| **Spearman Correlation**        | Measures **rank-based monotonic correlation** between two variables.                             | Robust to outliers; good for monotonic but non-linear relationships.                             | Use when relationships are monotonic but not linear.                                                      |
| **Kendallâ€™s Tau**               | Measures correlation between **ranked data** using concordant and discordant pairs.              | Better for small datasets; robust to ties in data.                                               | Use with ranked or ordinal data or when there are ties.                                                   |
| **Mutual Information (MI)**     | Captures **non-linear dependencies** based on information theory.                               | Detects both linear and non-linear relationships.                                                | Use when relationships are complex and non-linear.                                                        |
| **Normalized Mutual Information (NMI)** | Scaled version of MI to range [0, 1].                                                         | Makes MI values comparable across datasets.                                                      | Use when comparing dependencies across multiple variables.                                                |
| **Cross-Correlation Function (CCF)** | Measures correlation between two series at different **lags**.                                 | Useful for detecting lagged relationships in time series.                                        | Use when relationships may exist with a delay between series.                                             |
| **Granger Causality**           | Tests whether one time series can **predict another**. $H_0$: X doesn't predicts Y                                           | Identifies predictive relationships, even if not causal.                                         | Use to test whether lagged values of one series provide information about another.                        |
| **Dynamic Time Warping (DTW)**  | Measures similarity between two time series by **aligning them non-linearly**.                  | Good for detecting similarity in patterns even if series are shifted or misaligned in time.      | Use for similarity comparisons when time alignment issues exist.                                          |
| **Cointegration Tests**         | Tests whether two non-stationary series have a **long-term equilibrium relationship**. $H_0$: no co-integration         | Detects relationships even in non-stationary data.                                               | Use when series have trends but move together over time.                                                  |
| **Distance Correlation**        | Measures both linear and non-linear relationships between variables.                            | Captures dependencies beyond linear or monotonic.                                                | Use when the relationship form is unknown.                                                                |
| **Partial Correlation**         | Measures the correlation between two variables, **controlling for other variables**.             | Removes the effect of confounding variables.                                                     | Use when relationships may be confounded by a third variable.                                             |


## Key Additions for Long-Term Analysis

1. Spearman/Kendall Correlation:
    1.Captures the monotonic relationships over the entire scale, including trends and cycles.

2. Cross-Correlation (CCF):
    1. Analyzes how one series lags or leads the other, directly on the large scales.

3. Granger Causality:
    1. Identifies predictive relationships while preserving long-term trends.

4. Mutual Information:
    1. Captures non-linear dependencies while keeping the series intact.

5. Cointegration:
    1. Tests if two series move together in the long run, even if they have distinct trends.

6. Dynamic Time Warping (DTW):
    1. Aligns the global shapes of the time series, useful for comparing trends and cycles even when they occur at different times.

````python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, spearmanr, kendalltau
from sklearn.feature_selection import mutual_info_regression
from statsmodels.tsa.stattools import ccf, grangercausalitytests
from statsmodels.tsa.stattools import coint
from statsmodels.tsa.seasonal import STL
from dcor import distance_correlation
from pingouin import partial_corr
from fastdtw import fastdtw
import matplotlib.pyplot as plt
import seaborn as sns

# Generate example time series with trends and cycles
np.random.seed(42)
time = np.arange(100)
series1 = 0.5 * time + 10 * np.sin(0.2 * time) + np.random.normal(scale=2, size=len(time))  # Trend + cycle
series2 = 0.4 * time + 8 * np.sin(0.2 * time + np.pi/4) + np.random.normal(scale=2, size=len(time))  # Trend + shifted cycle

# Convert to DataFrame
df = pd.DataFrame({'Time': time, 'Series1': series1, 'Series2': series2})

# Plot the series
plt.figure(figsize=(10, 5))
plt.plot(df['Time'], df['Series1'], label='Series 1')
plt.plot(df['Time'], df['Series2'], label='Series 2')
plt.legend()
plt.title("Time Series")
plt.show()

# Detrend using STL decomposition (combine residuals and seasonal components)
detrended1 = series1 - pd.Series(series1).rolling(window=10).mean()
detrended2 = series2 - pd.Series(series2).rolling(window=10).mean()

plt.figure(figsize=(10, 5))
plt.plot(df['Time'], detrended1, label='Detrended Series 1')
plt.plot(df['Time'], detrended2, label='Detrended Series 2')
plt.legend()
plt.title("Detrended Time Series using STL")
plt.show()

# 1. Pearson Correlation
pearson_corr, _ = pearsonr(detrended1, detrended2)
print(f"Pearson Correlation (detrended with STL): {pearson_corr:.2f}")

# 2. Spearman Correlation
spearman_corr, _ = spearmanr(series1, series2)
print(f"Spearman Correlation: {spearman_corr:.2f}")

# 3. Kendall's Tau
kendall_corr, _ = kendalltau(series1, series2)
print(f"Kendall's Tau: {kendall_corr:.2f}")

# 4. Cross-Correlation Function (CCF)
cross_corr = ccf(detrended1, detrended2)
print(f"Cross-Correlation (lag 0): {cross_corr[0]:.2f}")

# 5. Granger Causality Test
data = pd.concat([pd.Series(series1), pd.Series(series2)], axis=1)
data.columns = ['Series1', 'Series2']
print("\nGranger Causality Test:")
grangercausalitytests(data, maxlag=2)
````

````python
# 6. Mutual Information
mi = mutual_info_regression(np.array(detrended1.dropna()).reshape(-1, 1), detrended2.dropna())
print(f"Mutual Information: {mi[0]:.2f}")

# 7. Cointegration Test
coint_t, p_value, _ = coint(series1, series2)
print(f"Cointegration Test p-value: {p_value:.2f}")

# 8. Distance Correlation
dist_corr = distance_correlation(series1, series2)
print(f"Distance Correlation: {dist_corr:.2f}")

# 9. Partial Correlation (control for time)
df['Time'] = time
partial_corr_result = partial_corr(data=df, x='Series1', y='Series2', covar='Time')
print(f"Partial Correlation (controlling for time): {partial_corr_result['r'].iloc[0]:.2f}")

# 10. Dynamic Time Warping (DTW)
distance, path = fastdtw(detrended1.dropna(), detrended2.dropna())
print(f"Dynamic Time Warping Distance: {distance:.2f}")
````

To determine which time series in `X` (exogenous variables) are useful for predicting those in `y` (endogenous variables), you can use **statistical tests**, **correlation analysis**, and **predictive modeling**. Below are the best tools and methods:

---

## **1ï¸âƒ£ Correlation & Dependence Analysis**
### âœ… **Pearson & Spearman Correlation**
- **Pearson**: Measures linear relationships.
- **Spearman**: Captures monotonic relationships (better for non-linear relationships in time series).
  
```python
# Compute Pearson correlation between all exogenous (X) and endogenous (y) time series
pearson_corr = y.corrwith(X, method="pearson")

# Compute Spearman correlation (for ranking-based dependencies)
spearman_corr = y.corrwith(X, method="spearman")

print("Pearson Correlation:\n", pearson_corr)
print("Spearman Correlation:\n", spearman_corr)
```
âœ” **Use when:** Looking for direct relationships.  
âš  **Limitation:** Doesnâ€™t capture lag effects.

---

### âœ… **Granger Causality Test**
- Checks **whether past values** of an exogenous series **help predict** the endogenous series.
- If p-value < 0.05, `X` **Granger-causes** `y`.

```python
from statsmodels.tsa.stattools import grangercausalitytests

max_lag = 5  # Choose based on domain knowledge
test_results = grangercausalitytests(pd.concat([y, X], axis=1), max_lag, verbose=True)
```
âœ” **Use when:** You suspect past values of `X` affect `y`.  
âš  **Limitation:** Assumes stationarity, sensitive to lag choice.

---

### âœ… **Cross-Correlation Function (CCF)**
- Measures the **correlation between `X` and `y`** across different lags.
  
```python
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import ccf

# Choose one column for visualization
exog_series = X.iloc[:, 0]  # First exogenous variable
endog_series = y.iloc[:, 0]  # First endogenous variable

cross_corr = ccf(endog_series, exog_series)

plt.plot(cross_corr)
plt.title("Cross-Correlation Function")
plt.show()
```
âœ” **Use when:** Finding the best lag for relationships.  
âš  **Limitation:** Works best with **stationary** series.

## **2ï¸âƒ£ Feature Importance Using Predictive Models**
### âœ… **Lasso Regression (Feature Selection)**
Lasso (L1-regularized regression) selects the most important predictors.

```python
from sklearn.linear_model import LassoCV
import numpy as np

# Ensure no NaN values
X_clean = X.dropna()
y_clean = y.dropna()

# Fit Lasso to find important variables
lasso = LassoCV(cv=5).fit(X_clean, y_clean)

# Show important variables
importance = np.abs(lasso.coef_)
selected_features = X_clean.columns[importance > 0]

print("Selected Exogenous Variables:\n", selected_features)
```
âœ” **Use when:** You want to automatically select key predictors.  
âš  **Limitation:** Assumes linear relationships.

---

### âœ… **Random Forest Feature Importance**
- Non-linear method to rank predictors.

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_clean, y_clean)

importance = rf.feature_importances_
important_features = X_clean.columns[np.argsort(-importance)]

print("Most Important Features:\n", important_features)
```
âœ” **Use when:** Relationships might be non-linear.  
âš  **Limitation:** Less interpretable than Lasso.

## **4ï¸âƒ£ Cointegration Tests (Long-Term Relationships)**
If two series move **together in the long run**, they are cointegrated.

### âœ… **Engle-Granger Cointegration Test**
```python

```
âœ” **Use when:** Checking **long-term** relationships.
